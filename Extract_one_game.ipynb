```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://store.steampowered.com/app/205270/3DMark_11/"
cookies = { 'birthtime': '283993201', 'mature_content': '1'}
r = requests.get(url, cookies = cookies)
html_doc = r.text
soup = BeautifulSoup(html_doc)
```

```python
print(soup)
```

```python
f = open("sample_file_2.txt", "w", encoding="utf_8")
f.write(html_doc)
f.close()
```

```python
#Game name
Game = soup.title.string[:-9]
```

```python
#Price
temp_result = soup.find_all(attrs={"itemprop":"price"})
Price = temp_result[0]['content']
print(Price)
```

```python
#Memory, Graphics, Storage
temp_result = soup.find_all('div', 'game_area_sys_req_rightCol')[0]
temp_result = temp_result.ul.select('ul > li')
Memory = 0
Graphics = ''
Storage = 0
for item in temp_result:
    item = item.get_text(strip=True)
    print(item)
    if 'Memory' in item:
        item = item.replace('Memory:', '')
        item = item.replace('RAM', '')
        item = item.split(' ')
        for w in item:
            try:
                Memory = float(w)
            except ValueError:
                pass
        if 'GB' not in item:
            Memory = Memory/1024
    elif 'Graphics' in item:
        Graphics = item[9:]
    elif 'Storage' in item:
        item = item.replace('Storage:', '')
        item = item.split(' ')
        for w in item:
            try:
                Storage = float(w)
            except ValueError:
                pass
        if 'GB' not in item:
            Storage = Storage/1024
```

```python
#Game tag
temp_result = soup.find_all('a', 'app_tag')
Tags = []
for item in temp_result:
    tag = item.get_text(strip=True)
    Tags.append(tag)
```

```python
#Release date
temp_result = soup.find_all('div', 'date')
Release = temp_result[0].get_text(strip=True)
```

```python
#Reviews
temp_result = soup.find_all('span', 'game_review_summary positive')
s = ''
for item in temp_result:
    try:
        if '30 days' not in item['data-tooltip-text']:
            s = item['data-tooltip-text']
    except KeyError:
        pass
s = s.replace(' user reviews for this game are positive.', '')
s = s.replace('of the ', '')
s = s.replace(',', '')
Pos_reviews, No_reviews = s.split(' ')
Pos_reviews = float(Pos_reviews[:-1])
No_reviews = int(No_reviews)
```

```python
df.loc[0] = [Game, Price, Memory, Graphics, Storage, Tags, Release, No_reviews, Pos_reviews]
```

```python
class Browsing:
    def __init__(self):
        self.url_list = []

    def GetUrls(self, url):
        r = requests.get(url)
        html_doc = r.text
        soup = BeautifulSoup(html_doc)
        temp_result = soup.find_all('a', 'search_result_row ds_collapse_flag ')
        urls = []
        for item in temp_result:
            urls.append(item['href'])
        return urls

    def GetAllUrls(self):
        root_search_url = 'https://store.steampowered.com/search/?'
        url_list = self.GetUrls(root_search_url)
        for i in range(2, 10):
            new_urls = []
            new_search_url = root_search_url + 'page=' + str(i)
            new_urls = self.GetUrls(new_search_url)
            url_list = [*url_list, *new_urls]
        return url_list
    
B = Browsing()
t = B.GetAllUrls()
print(t)
```
